## Data Manipulation with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,3,4,5,8,9]
y=[2,3,4,6,7,9]
plt.plot(x,y)
plt.xlabel("Its an X Axis")
plt.ylabel("Its an Y Axis")
plt.title("Sample Graph")
```

#### Plotting An Customised Plot

```python
x=[1,3,4,5,8,9]
y=[2,3,4,6,7,9]
plt.plot(x,y,color="red",linestyle="--",marker="*",linewidth="4",markersize="9")
```

#### Multiple Plots

```python
a = [1,3,5,7,8,]
b = [1,2,4,6,8,]
c = [2,4,6,8,9,]

plt.figure(figsize=(15,9))

plt.subplot(1,3,1)
plt.plot(a,b,color='green')
plt.title('Plot 1')

plt.subplot(1,3,2)
plt.plot(a,c,color='green')
plt.title('Plot 2')

plt.subplot(1,3,3)
plt.plot(c,b,color='green')
plt.title('Plot 3')
```

#### Bar Plots

```python
categories = ['A','B','C','D','F']
valuews = [12,14,7,15,11]
plt.bar(categories,valuews,color=("yellow","green","Purple"))        # We Can give A Multiple Colors  
plt.xlabel("Its An X Label")
plt.title("Bar Chart")
```

#### Histograms


```python
numbers = [2,3,3,4,5,4,2,1,2,2,2,3,4,2,1,3,4,4]
plt.hist(numbers,bins=5,color=("Orange"),edgecolor=("Black"))        # Here The bins is the number of bins we want to see in the histogram
```
