## Data Visualization Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

#### Scatterplot

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatteer plot of tips')
```

#### Line plot

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title('Line plot of tips')  
```

#### Categorical Bar Chart

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

#### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
### This plot Will Helps to Find The outliers as well as the distribution of the data. with Min Nad Max values
```

#### Histograms

```python
sns.histplot(tips['total_bill'],bins=20,kde=True)               # Kde(kernal density estimator) Gives the Curve line
```

#### HeatMap

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
